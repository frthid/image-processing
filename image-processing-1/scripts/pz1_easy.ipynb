```python
url = "/content/drive/MyDrive/image-processing/image-processing-1/assets/hedgehog.jpg"
```

```python
image = io.imread(url)
```

```python
plt.imshow(image)
plt.show()
```

```python
image.shape
```

```python
noise = np.random.normal(0, 15, image.shape[:2])
filter_image = np.clip(image + noise[:, :, np.newaxis], 0, 255).astype(np.uint8)

filter_image = cv2.cvtColor(filter_image, cv2.COLOR_RGB2GRAY)

brightness_factor = 1.2
filter_image = np.clip(filter_image * brightness_factor, 0, 255).astype(np.uint8)

plt.figure(figsize=(18, 6))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Исходное изображение")

plt.subplot(1, 2, 2)
plt.imshow(filter_image, cmap="gray")
plt.title("Измененное")
```
